## Setup

Load libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
from yellowbrick.cluster import KElbowVisualizer
```

```python
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
```

## Load dataset

```python
link_to_dataset='https://archive.ics.uci.edu/ml/machine-learning-databases/00267/data_banknote_authentication.txt'
col_names=['variance', 'skewness', 'kurtosis', 'entropy', 'class']
banknote=pd.read_csv(link_to_dataset, header=None, names=col_names)
banknote.info()
banknote.head()
```

## Visualization

```python
sns.histplot(data=banknote, x='variance')
plt.show()
```

```python
plt.figure(figsize=(15,8))

for plot in range(4):
    plt.subplot(2, 2, plot+1)
    sns.histplot(data=banknote, x=banknote.iloc[:, plot])
plt.show()
```

### Dimensionality Reduction

Dimension reduction reduces the number of varables to the desired count. We'll plot using t-SNE.

```python
X= banknote.drop('class', axis=1)
```

### Scale variables


```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

### t-SNE

```python
reducer=TSNE(n_components=2, random_state=0)
reduced=reducer.fit_transform(X)
tsne_components=pd.DataFrame(reduced, columns=['tsne1', 'tsne2'])
tsne_components.head()
```

```python
sns.scatterplot(x=tsne_components.tsne1, y=tsne_components.tsne2, edgecolor='black')
```

```python
pca=PCA(n_components=2)
reduced=pca.fit_transform(X_scaled)
pca_components=pd.DataFrame(reduced, columns=['PCA1', 'PCA2'])
pca_components.head()
```

```python
sns.scatterplot(x=pca_components.PCA1, y=pca_components.PCA2, edgecolor='black')
```

### kMeans

```python
model=KMeans()
visualizer=KElbowVisualizer(model, k=(2, 10))

visualizer.fit(X)
visualizer.show()
```

```python
model=KMeans()
visualizer=KElbowVisualizer(model, k=(2, 10))

visualizer.fit(X_scaled)
visualizer.show()
```

```python
model=KMeans()
visualizer=KElbowVisualizer(model, k=(2, 10), metric='silhouette')

visualizer.fit(X_scaled)
visualizer.show()
```

```python
model=KMeans()
visualizer=KElbowVisualizer(model, k=(2, 10), metric='calinski_harabasz')

visualizer.fit(X_scaled)
visualizer.show()
```

## kMeans clustering proper

```python
cluster=KMeans(n_clusters=2, random_state=0)
clusters=cluster.fit_predict(X_scaled)
```

```python
plt.figure(figsize=(15,6))

plt.subplot(121)
sns.scatterplot(x=pca_components.PCA1, y=pca_components.PCA2, edgecolor='black', hue=clusters)
plt.title('KMeans')

plt.subplot(122)
sns.scatterplot(x=pca_components.PCA1, y=pca_components.PCA2, edgecolor='black', hue=banknote['class'])
plt.title('True Value')

plt.show()
```

### Heirarchical Clustering

```python
hei = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
clustershe=hei.fit_predict(X_scaled)
```

```python
np.unique(clustershe, return_counts=True)
```

```python
plt.figure(figsize=(15,5))

plt.subplot(131)
sns.scatterplot(x=pca_components.PCA1, y=pca_components.PCA2, edgecolor='black', hue=clusters)
plt.title('KMeans')

plt.subplot(132)
sns.scatterplot(x=pca_components.PCA1, y=pca_components.PCA2, edgecolor='black', hue=clustershe)
plt.title('Heirarchical')

plt.subplot(133)
sns.scatterplot(x=pca_components.PCA1, y=pca_components.PCA2, edgecolor='black', hue=banknote['class'])
plt.title('True Value')

plt.show()
```

```python
sil_kmean=metrics.silhouette_score(X, clusters)
sil_hei=metrics.silhouette_score(X, clustershe)

print(f'Silhouette Score\n=================')
print(f'KMeans: {sil_kmean}')
print(f'Heirarchical: {sil_hei}')
```

```python
sil_kmean
```

```python
print(metrics.homogeneity_score(banknote['class'].values, clusters))
print(metrics.completeness_score(banknote['class'].values, clusters))
```

## Another Example

```python
link1='https://archive.ics.uci.edu/ml/machine-learning-databases/mfeat/mfeat-fac'
link2='https://archive.ics.uci.edu/ml/machine-learning-databases/mfeat/mfeat-fou'

data1=pd.read_csv(link1, header=None, delim_whitespace=True)
data2=pd.read_csv(link2, header=None, delim_whitespace=True)
```

```python
data1.head()
```

```python
data2.head()
```

```python
final_data=pd.concat([data1, data2], ignore_index=True, axis=1)
```

```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(final_data)
```

```python
reducer=TSNE(n_components=2, random_state=42)
reduced=reducer.fit_transform(X_scaled)
tsne_components=pd.DataFrame(reduced, columns=['tsne1', 'tsne2'])
tsne_components.head()
```

```python
y=np.repeat(range(0,10), 200)
```

```python
sns.scatterplot(x=tsne_components.tsne1, y=tsne_components.tsne2, edgecolor='black')
```

```python
cluster=KMeans(n_clusters=10, random_state=42)
clusters=cluster.fit_predict(X_scaled)
```

```python
plt.figure(figsize=(15,6))

plt.subplot(121)
sns.scatterplot(x=tsne_components.tsne1, y=tsne_components.tsne2, edgecolor='black', hue=clusters, palette='tab10')
plt.title('KMeans')

plt.subplot(122)
sns.scatterplot(x=tsne_components.tsne1, y=tsne_components.tsne2, edgecolor='black', hue=y, palette='tab10')
plt.title('True Value')

plt.show()
```

```python
hei = AgglomerativeClustering(n_clusters=10, affinity='euclidean', linkage='ward')
clustershe=hei.fit_predict(X_scaled)
```

```python
plt.figure(figsize=(15,5))

plt.subplot(131)
sns.scatterplot(x=tsne_components.tsne1, y=tsne_components.tsne2, edgecolor='black', hue=clusters, palette='tab10')
plt.title('KMeans')

plt.subplot(132)
sns.scatterplot(x=tsne_components.tsne1, y=tsne_components.tsne2, edgecolor='black', hue=clustershe, palette='tab10')
plt.title('Heirarchical')

plt.subplot(133)
sns.scatterplot(x=tsne_components.tsne1, y=tsne_components.tsne2, edgecolor='black', hue=y, palette='tab10')
plt.title('True Value')

plt.show()
```

```python
sil_kmean=metrics.silhouette_score(final_data, clusters)
sil_hei=metrics.silhouette_score(final_data, clustershe)

print(f'Silhouette Score\n=================')
print(f'KMeans: {sil_kmean}')
print(f'Heirarchical: {sil_hei}')
```

```python
print(metrics.homogeneity_score(y, clusters))
print(metrics.completeness_score(y, clusters))
```

```python
print(metrics.homogeneity_score(y, clustershe))
print(metrics.completeness_score(y, clustershe))
```

```python
def evaluation(y_true, y_fitted, X, model_name):
    
    Homo=metrics.homogeneity_score(y_true, y_fitted)
    Comp=metrics.completeness_score(y_true, y_fitted)
    Vmea=metrics.v_measure_score(y_true, y_fitted)
    AdjR=metrics.adjusted_rand_score(y_true, y_fitted)
    AdjM=metrics.adjusted_mutual_info_score(y_true, y_fitted)
    sil=metrics.silhouette_score(X, y_fitted)
    
    metric_value=[Homo, Comp, Vmea, AdjR, AdjM, sil]
    index=[
        'Homogeneity', 'Completeness', 'V-measure',
        'Adj. Rand Index', 'Adj. Mutual Information', 'Silhouette Score' 
    ]
    
    results=pd.DataFrame({
        model_name: metric_value
    }, index=index)
    
    return results
```

```python
kmean_rslt=evaluation(y, clusters, X_scaled, 'KMeans')
hei_rslt=evaluation(y, clustershe, X_scaled, 'Heirarchical')

performance=pd.concat([kmean_rslt, hei_rslt], axis=1)
performance
```

```python
performance.plot(kind='bar', figsize=(7, 5), fontsize=12)
plt.xlabel('Evaluation Metric')
plt.ylabel('Score')
plt.grid(False)
plt.show()
```
